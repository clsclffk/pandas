```python
tips = sns.load_dataset('tips')
tips
```

## 일변량 그래프
- 하나의 변수에 대한 데이터 시각화
1. 히스토그램
2. 막대그래프
3. 박스플롯

```python
fig = plt.figure()
axes = fig.add_subplot(1,1,1)
axes.hist(data=tips, x='total_bill', bins=20)
plt.show()
```

```python
fig = plt.figure()
axes = fig.add_subplot(1,1,1)
sns.countplot(data=tips, x='day', ax=axes, palette='pastel')
plt.show()
```

```python
fig = plt.figure()
axes = fig.add_subplot(1,1,1)
axes.boxplot(data=tips, x='total_bill')
plt.show()
```

## 이변량 그래프
1. 산점도
2. 박스플롯

```python
fig = plt.figure()
axes = fig.add_subplot(1,1,1)
axes.scatter(data=tips, x='total_bill', y='tip')
plt.show()
```

```python
fig = plt.figure()
axes = fig.add_subplot(1,1,1)
axes.boxplot(x=[tips[tips['sex']=='Female']['tip'], tips[tips['sex']=='Male']['tip']],
             labels=['Female','Male']
)
plt.show()
```

## 다변량 그래프

```python
#map()함수에 사용
colors = {"Female": "#f1a340", "Male": "#998ec3"} # 성별에 따른 색상을 지정, 여성은 주황색, 남성은 보라색 
#16진수 코드는 https://www.w3schools.com/colors/colors_picker.asp 에서 확인 가능

scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1, 1, 1)

axes1.scatter(data=tips,
              x='total_bill',
              y='tip',
              s=tips['size']**2*10,# size를 제곱하여 점의 크기를 키움, s는 점의 크기를 나타내는 매개변수
              c=tips['sex'].map(colors),# c는 점의 색을 나타내는 매개변수
              alpha=0.5) 

axes1.set_title('Colored by Sex and Sized by Size')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')

scatter_plot.suptitle('Total Bill vs Tip')

plt.show()
```

## 이변량 그래프

```python
fig, ax = plt.subplots()
sns.scatterplot(data=tips, x='total_bill', y='tip', ax=ax, hue='day')
ax.set_title('지불금액과 팁의 관계')
plt.show()
```

```python
fig, ax = plt.subplots()
sns.regplot(data=tips, x='total_bill', y='tip', ax=ax)
ax.set_title('지불금액과 팁의 관계')
plt.show()
```

```python
x = tips['total_bill']
y = tips['tip']

# 회귀선의 기울기와 절편 계산
slope, intercept = np.polyfit(x, y, 1)

# 회귀선 출력
print(f"기울기: {slope}, 절편: {intercept}")
```

```python
j = sns.jointplot(data=tips, x='total_bill', y='tip')
j.set_axis_labels(xlabel='TotalBill', ylabel='Tip')
j.figure.suptitle('jointplot')
plt.show()
```

```python
kde, ax = plt.subplots()
sns.kdeplot(data=tips, x='total_bill', y='tip', fill=True, ax=ax)
ax.set_title('KDE plot')
plt.show()
```

```python
bar, ax = plt.subplots()
sns.barplot(data=tips, x='time', y='total_bill', ax=ax, estimator=np.mean, errorbar=None)
ax.set_title('barplot')
plt.show()
```

```python
fig, ax = plt.subplots()
sns.boxplot(data=tips, x='time', y='total_bill', ax=ax)
ax.set_title('boxplot')
plt.show()
```

```python
fig, ax = plt.subplots()
sns.violinplot(data=tips, x='time',y='total_bill', ax=ax)
ax.set_title('Violinplot')
plt.show()
```

```python
fig = sns.pairplot(data=tips)
fig.figure.suptitle('tips 데이터 관계분석')
plt.tight_layout()
plt.show()
```

```python
g = sns.FacetGrid(tips, col="time")
g.map(sns.violinplot, "total_bill")
g.set_titles(col_template="{col_name} 시간")
plt.show()
```
