## Series & DataFrame 만들기
1. `pd.Series()`

```python
pandas.Series(
    data=None,         # 시리즈에 입력할 데이터 (리스트, 배열, 사전 등)
    index=None,        # 시리즈의 인덱스 (기본값은 0부터 시작하는 정수 인덱스)
    dtype=None,        # 시리즈의 데이터 타입 (기본값은 자동 결정)
    name=None,         # 시리즈의 이름
    copy=False,        # 입력 데이터를 복사할지 여부 (기본값 False)
    fastpath=False     # 내부적으로 사용하는 매개변수 (일반적으로 사용하지 않음)
)
```

2. `pd.DataFrame()`

```python
pandas.DataFrame(
    data=None,         # 데이터프레임에 입력할 데이터 (리스트, 사전, 배열 등)
    index=None,        # 행 인덱스 (기본값은 0부터 시작하는 정수 인덱스)
    columns=None,      # 열 이름 (기본값은 자동 결정)
    dtype=None,        # 데이터프레임의 데이터 타입 (기본값은 자동 결정)
    copy=False         # 입력 데이터를 복사할지 여부 (기본값 False)
)```

    keys,               # 인덱스로 설정할 열(컬럼) 이름 또는 이름 리스트

    append=False,       # 기존 인덱스에 추가할지 여부 (기본값 False)
    inplace=False,      # 원본 DataFrame을 수정할지 여부 (기본값 False, 새로운 DataFrame 반환)
    verify_integrity=False  # 인덱스의 유일성을 강제할지 여부 (기본값 False)
)

````python
# 직원 명단 DataFrame 생성

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [25, 45, 35, 32, 28],
    'Salary': [5000, 7000, 6500, 5500, 6000]
}

df = pd.DataFrame(data)
df
````

````python
# 나이에 따라 직원의 나이를 분류하는 함수
# young, mid, experience
def age(x):
    if x // 10 == 2:
        return 'young'
    elif x // 10 == 3:
        return 'mid'
    else:
        return 'experience'
````

````python
# Category 칼럼 추가
df['Category'] = df['Age'].apply(age)
df
````

````python
# salary에 보너스 10%를 증가한 칼럼 Bonus칼럼 추가
df['Bonus'] = df['Salary'].apply(lambda x : x * 1.1)
df
````

## 변환
- 표준화(Z-score) : 데이터의 평균을 0, 표준편차를 1로 변환
  - 아웃라이어에 강함
    
- 정규화(Min-Max Scaling) : 데이터의 범위를 [0, 1] 또는 [-1, 1]로 변환
  - 아웃라이어의 영향받음

````python
def my_zscore(x):
    return (x - x.mean()) / x.std()
````

````python
df.groupby('continent').lifeExp.transform(my_zscore)
````
