## Series & DataFrame 만들기
1. `pd.Series()`

```python
pandas.Series(
    data=None,         # 시리즈에 입력할 데이터 (리스트, 배열, 사전 등)
    index=None,        # 시리즈의 인덱스 (기본값은 0부터 시작하는 정수 인덱스)
    dtype=None,        # 시리즈의 데이터 타입 (기본값은 자동 결정)
    name=None,         # 시리즈의 이름
    copy=False,        # 입력 데이터를 복사할지 여부 (기본값 False)
    fastpath=False     # 내부적으로 사용하는 매개변수 (일반적으로 사용하지 않음)
)
```

2. `pd.DataFrame()`

```python
pandas.DataFrame(
    data=None,         # 데이터프레임에 입력할 데이터 (리스트, 사전, 배열 등)
    index=None,        # 행 인덱스 (기본값은 0부터 시작하는 정수 인덱스)
    columns=None,      # 열 이름 (기본값은 자동 결정)
    dtype=None,        # 데이터프레임의 데이터 타입 (기본값은 자동 결정)
    copy=False         # 입력 데이터를 복사할지 여부 (기본값 False)
)```

## 특정 열을 인덱스로 설정

1. `set_index()`

```python
DataFrame.set_index(
    keys,               # 인덱스로 설정할 열(컬럼) 이름 또는 이름 리스트
    drop=True,          # 기존의 열을 DataFrame에서 삭제할지 여부 (기본값 True)
    append=False,       # 기존 인덱스에 추가할지 여부 (기본값 False)
    inplace=False,      # 원본 DataFrame을 수정할지 여부 (기본값 False, 새로운 DataFrame 반환)
    verify_integrity=False  # 인덱스의 유일성을 강제할지 여부 (기본값 False)
)
```

## 날짜 변환
1. `to_datetime()`
2. `'dt' 속성`
3. `apply()`
4. `assign()`

````python
df.info()
````

````python
born_datetime = pd.to_datetime(df['Born'], format='%Y-%m-%d')
born_datetime
````

````python
df['born_dt'] = born_datetime
df.head()
````

````python
df['died_dt'] = pd.to_datetime(df['Died'], format='%Y-%m-%d')
````

````python
df['age_days'] = df['died_dt'] - df['born_dt']
df['age_days'].dtype
````

````python
df['age_days'].dt.days / 365
````

````python
import numpy as np

(df['age_days'].dt.days / 365).apply(np.floor)
````

````python
df['ages_days'] = (df['age_days'].dt.days / 365).apply(np.floor)
df
````
